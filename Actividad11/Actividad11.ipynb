```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
from math import cos
```

```python
def f(t, z, omega):
    
    #Right hand side of the differential equations
    #  dx/dt = y
    #  dy/dt = omega * x
    
    δ, α, β, γ, ω = omega                       #Parámetros de la integral
    x, y = z                                    #Parámetros de integración
    f = [y, - δ*y - α*x - β*x**3 + γ*cos(ω*t)]  #Función a integrar

    return f
```

```python
solver = ode(f)
solver.set_integrator('dop853')
```

```python
w1 = []
t1 = 80
n = 600

for i in range(0, n):
    
    γ = 0.20
    α = -1.0        
    δ = 0.3
    ω = 1.2
    β = 1.0

    val = δ, α, β, γ, ω                      
    solver.set_f_params(val)
    
    t0 = 0.0
    z0 = [1.0, 0]
    solver.set_initial_value(z0, t0)

    t = np.linspace(t0, t1, n)
    sol = np.empty((n, 2))
    sol[0] = z0

    k = 1

    while solver.successful() and solver.t < t1:
        solver.integrate(t[k])
        sol[k] = solver.y
        k += 1

    x = sol[:,0]
    y = sol[:,1]

    w1.append(ω)

T = []
for i in range (0, n):
    k = (t[i]) * (w1[i]) / (2 * np.pi)
    T.append(k)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(T, sol[:,0], label = 'x')
plt.xlabel('t/T')
plt.ylabel('x')
plt.title("γ = 0.20")
plt.grid(True)
plt.legend()
plt.savefig("duff20.png")
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(sol[:,0], sol[:,1], label = 'x', marker = ".")
plt.xlabel('x')
plt.ylabel('dx/dt')
plt.title("γ = 0.20")
plt.grid(True)
plt.legend()
plt.savefig("poin20.png")
plt.show()
```
