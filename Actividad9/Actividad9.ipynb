```python
def vectorfield(w, t, p):
    x1, y1, x2, y2 = w
    m1, m2, k1, k2, k3, L1, L2, b1, b2 = p
    
    f = [y1,
         (-b1 * y1 - k1 * (x1 - L1) - k2 * ( x1 - L1 + L2 - x2 )) / m1,
         y2,
         (-b2 * y2 - k3 * ( x2 - L2 ) - k2 * ( x2 - L2 + L1 - x1 ))/ m2]  
    return f
```

```python
open('dos_masas.txt', 'w').close()
from scipy.integrate import odeint
#Masas
m1 = 1.0
m2 = 1.0
#Constantes de los resortes
k1 = 1.2
k2 = 1.2
k3 = 1.2
#Coeficientes de fricci칩n
b1 = 0.1
b2 = 0.2
#Longitud natural
L1 = 0.5
L2 = 1.0
#Condiciones iniciales, x1 y x2 son las posiciones iniciales y y1, y2 las velocidades inciales
x1 = 1.0
y1 = 1.0
x2 = 0.0
y2 = 0.0

#Par치metros de resoluci칩n de EDO (Ecuaci칩n Diferencial Ordinaria)
abserr = 1.0e-8
relerr = 1.0e-6
stoptime = 30.0
numpoints = 750

#Ancho de paso
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

p = [m1, m2, k1, k2, k3, L1, L2, b1, b2]
w0 = [x1, y1, x2, y2]

#Llamamos al solucionador de EDO
wsol = odeint(vectorfield, w0, t, args = (p,), atol = abserr, rtol = relerr)

for t1, w1 in zip(t, wsol):
    print (t1, w1[0], w1[1], w1[2], w1[3],file = open('dos_masas.txt','a'))
```
